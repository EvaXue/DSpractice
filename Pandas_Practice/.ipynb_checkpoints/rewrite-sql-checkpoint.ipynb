```python
airports = pd.read_csv('data/airports.csv')
airport_freq = pd.read_csv('data/airport-frequencies.csv')
runways = pd.read_csv('data/runways.csv')
```

```python
airports.loc[airports.ident == 'KLAX']
airports.type.unique()
airports.loc[(airports.iso_region == 'US-CA')&(airports.type == 'seaplane_base')]
airports.loc[(airports.iso_region == 'US-CA')&(airports.type == 'large_airport'),['ident', 'name', 'municipality']]
airport_freq.loc[(airport_freq.airport_ident == 'KLAX' )].sort_values('type', ascending =False)
airports.loc[~airports.type.isin(['heliport', 'balloonport'])]
airports.groupby(['iso_country','type']).size().to_frame('count').reset_index().sort_values(['iso_country','type'], ascending = [True,False])
```
